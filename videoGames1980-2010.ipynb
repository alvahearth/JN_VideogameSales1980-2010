## Asking and Answering Questions

## 1.- Which are the 3 most popular genres of videogames ?

```python
vg_genre = vg_df.groupby(["Genre"]).count()
```

```python
sample_1 = vg_genre["Name"]
```

```python
plt.figure(figsize=(15,5))
plt.plot(sample_1,"H-b");
```

```python
sample_1.nlargest(n=3)
```

### We can conclude with this data, that the 3 most popular genres are Action games, Sports games and Misc games

#### Note: "Misc" games refers to any game that does not fit into the more traditional genres, for example mobile games, family games, advertising games and the like.

## 2.- In which year was the peak of videogames sold?

```python
plt.figure(figsize=(16,4))
plt.title("Videogames sales throughout the years")
plt.xlabel("Years")
plt.ylabel("Sales (Per Millions)")
plt.plot(vg_year.index,vg_year.loc[:,"Global_Sales"],"o-b")
plt.bar(vg_year.index,vg_year.loc[:,"Global_Sales"]);
```

```python
vg_year["Global_Sales"].nlargest(n=1)
```

### The peak of videogames sales was in 2008 with more than 678 millions

## 3.- What are the 10 best selling game according to this data

```python
best_selling_vg = vg_df.sort_values(by="Global_Sales",ascending=False).head(10)
```

```python
best_selling_vg
```

```python
plt.figure(figsize=(24,5))
plt.xlabel("Videogames")
plt.ylabel("Units Sold")
plt.title("The 10 best selling games up to 2010")
plt.bar(best_selling_vg["Name"],best_selling_vg["Global_Sales"]);
```

### These are the top 10 videogames sold until 2010, as you can see the sales of wii sports are twice the amount of Super Mario Brothers, the second game in the list, this shows that games that are not too competive or complex are really popular.

### And we can also infer from this chart is that all the games were published by Nintendo.

## 4.- Which videogame platform has the most games released

```python
vg_games = vg_df.groupby(by="Platform").count()
```

```python
vg_games
```

```python
plt.figure(figsize=(18,5))
plt.plot(vg_games,"H-b");
```

### The Playstation 2 has the largest library of games with more than 2100, followed closely by the Nintendo DS with 1944 games

## 5.- How the sales of videogames has changed troughout the years in the three main regions?

```python
plt.figure(figsize=(16,7))
plt.xlabel("Years")
plt.ylabel("Sales")
plt.bar(vg_year.index,vg_year.loc[:,"NA_Sales"],label="NA_Sales",alpha=0.5)
plt.bar(vg_year.index,vg_year.loc[:,"EU_Sales"],label="EU_Sales",alpha=0.5)
plt.bar(vg_year.index,vg_year.loc[:,"JP_Sales"],label="JP_Sales",alpha=0.5)
plt.legend();
```

### The sales of videogames in the early 80's were dominated by the NA region, later in the mid 80's the JP region enter in the business with a heavy swing, with the consoles like the NES, The EU region play little part in the videogames business, while they had moderately high sales they were always below the margins of NA and JP, that is until 1995 when they started to expand more and more, as the market of videogames began to mature the population began to accept videogames as a "hobby", moving forward to the 2000's and beyond, we can see that the NA and EU have had the lead in sales over JP since then, and we can clearly see a trend that seems to not slow down.
